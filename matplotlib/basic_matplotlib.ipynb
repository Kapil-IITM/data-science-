```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
mylabel = ['apple','mango','grapes','banana']
myexplode = [0, 0.1, 0, 0]  # Slightly explode the 'mango' slice

plt.figure(figsize=(8,8))  # Larger figure for better appearance
plt.pie(
    x,
    labels=mylabel,
    startangle=90,
    explode=myexplode,
    shadow=True,
    autopct='%1.1f%%',
    colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'],
    textprops={'fontsize': 14}
)
plt.title('Fruit Distribution', fontsize=18)
plt.show()
```
